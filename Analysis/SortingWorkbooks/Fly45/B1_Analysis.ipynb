```python
sp = fly.processed_signals['spike_pool']
```

```python
null_mask = ones_like(sp)
```

```python
pca_0 = srtr.PCATransform(sp,null_mask,defaults.pca_trans)
pca_0.transform()
km_0 = srtr.KMeansCluster(sp,null_mask,pca_0.trnsmtrx,defaults.km_cluster)
km_0.select()
sview.plot_clusters(sp,km_0)
display(gcf())
```

```python
mask_00 = km_0.mask_from_labels('0')
mask_01 = km_0.mask_from_labels('1')
sview.plot_select_trains(sp,[mask_00,mask_01])
```

```python
pca_01 = srtr.PCATransform(sp,mask_01,defaults.pca_trans)
pca_01.transform()
km_01 = srtr.KMeansCluster(sp,mask_01,pca_01.trnsmtrx,defaults.km_cluster)
km_01.select()
sview.plot_clusters(sp,km_01)
display(gcf())
```

```python
mask_010 = km_01.mask_from_labels('0')
mask_011 = km_01.mask_from_labels('1')
sview.plot_select_trains(sp,[mask_00,mask_010,mask_011])
```

```python
md_params = {'trans_dims':1}
mdt_010 = srtr.MedTrans(sp,mask_010,md_params)
mdt_010.transform()
th_010 = srtr.ThreshSelector(sp,mask_010,mdt_010.trnsmtrx,{})
thresh = percentile(mdt_010.trnsmtrx,80)
th_010.select(thresh)
sview.plot_clusters(sp,th_010)
display(gcf())
```

```python
mask_0100 = th_010.mask_from_labels('0')
mask_0101 = th_010.mask_from_labels('1')
sview.plot_select_trains(sp,[mask_010,mask_0100,mask_0101])
```

```python
sp.annotations['spk_mask']  = mask_0100
```
