```python
fly.pool_params = pp
fly.extract_spike_pool()
fly.process_wb_signals()
```

```python
sp = fly.processed_signals['spike_pool']
```

```python
null_mask = ones_like(sp)
```

```python
p2p_0 = srtr.P2PTransform(sp,null_mask,{'trans_dims':1})
p2p_0.transform()
km_settings = {'kmeans_nc':2,'init':'k-means++'}
km_0 = srtr.KMeansCluster(sp,null_mask,p2p_0.trnsmtrx,km_settings)
km_0.select()
#hist(p2p_0.trnsmtrx[:,1],bins = 100)
sview.plot_clusters(sp,km_0)
display(gcf())
```

```python
mask_00 = km_0.mask_from_labels('0')
mask_01 = km_0.mask_from_labels('1')
sview.plot_select_trains(sp,[mask_00,mask_01])
```

```python
pca_00 = srtr.PCATransform(sp,mask_00,defaults.pca_trans)
pca_00.transform()
km_00 = srtr.KMeansCluster(sp,mask_00,pca_00.trnsmtrx,defaults.km_cluster)
km_00.select()
sview.plot_clusters(sp,km_00)
display(gcf())
```

```python
mask_001 = km_00.mask_from_labels('1')
mask_000 = km_00.mask_from_labels('0')
sview.plot_select_trains(sp,[mask_00,mask_01,mask_000])
```

```python
md_params = {'trans_dims':1}
mdt_000 = srtr.MedTrans(sp,mask_000,md_params)
mdt_000.transform()
th_000 = srtr.ThreshSelector(sp,mask_000,mdt_000.trnsmtrx,{})
thresh = percentile(mdt_000.trnsmtrx,98)
th_000.select(thresh)
sview.plot_clusters(sp,th_000)
display(gcf())
```

```python
mask_0000 = th_000.mask_from_labels('0')
mask_0001 = th_000.mask_from_labels('1')
sview.plot_select_trains(sp,[mask_00,mask_000,mask_0000,mask_0001])
```

```python
sp.annotations['spk_mask'] = mask_0000
```
