```python
fly.extract_spike_pool()
fly.process_wb_signals()
```

```python
sp = fly.processed_signals['spike_pool']
phases = fly.processed_signals['wb_phase']
times = fly.signals['LeftWing'].times
fly.processed_signals['spk_phases'] = stools.calc_spike_phases(sp,phases,times)
fly.processed_signals['spk_wbs'] = stools.calc_spike_wbs(sp,fly.processed_signals['wb_flips'],times)
```

```python
null_mask = ones_like(sp)
```

```python
pca_0 = srtr.PCATransform(sp,null_mask,defaults.pca_trans)
pca_0.transform()
km_settings = {'kmeans_nc':2,'init':'k-means++'}
km_0 = srtr.KMeansCluster(sp,null_mask,pca_0.trnsmtrx,km_settings)
km_0.select()
sview.plot_clusters(sp,km_0)
display(gcf())
```

```python
mask_00 = km_0.mask_from_labels('0')
mask_01 = km_0.mask_from_labels('1')
sview.plot_select_trains(sp,[mask_00,mask_01])
```

```python
mdt_01 = srtr.MedTrans(sp,null_mask,{'trans_dims':1})
mdt_01.transform()
th_01 = srtr.ThreshSelector(sp,null_mask,mdt_01.trnsmtrx,km_settings)
thresh = 280 #np.percentile(mdt_00.trnsmtrx,90)
th_01.select(thresh)
sview.plot_clusters(sp,th_01)
display(gcf())
```
