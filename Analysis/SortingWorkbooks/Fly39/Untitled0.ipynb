```python
pp = {'spk_pool_thresh':8.0,
               'spk_window_left' : 25,
               'spk_window_right' : 20,
               'spk_pool_filter_window' : 51}
tp = stools.get_spike_pool(test_sweep,
                      thresh = pp['spk_pool_thresh'],
                      wl = pp['spk_window_left'],
                      wr = pp['spk_window_right'],
                      filter_window = pp['spk_pool_filter_window'])
```

```python
fly.pool_params = pp
```

```python
fly.extract_spike_pool()
fly.process_wb_signals()
```

```python
sp = fly.processed_signals['spike_pool']
```

```python
null_mask = ones_like(sp)
```

```python
p2p_0 = srtr.P2PTransform(sp,null_mask,{'trans_dims':1})
p2p_0.transform()
km_settings = {'kmeans_nc':2,'init':'k-means++'}
km_0 = srtr.KMeansCluster(sp,null_mask,p2p_0.trnsmtrx,km_settings)
km_0.select()
#hist(p2p_0.trnsmtrx[:,1],bins = 100)
sview.plot_clusters(sp,km_0)
display(gcf())
```

```python
mask_00 = km_0.mask_from_labels('0')
pca_00 = srtr.PCATransform(sp,mask_00,defaults.pca_trans)
pca_00.transform()
km_00 = srtr.KMeansCluster(sp,mask_00,pca_00.trnsmtrx,defaults.km_cluster)
km_00.select()
sview.plot_clusters(sp,km_00)
display(gcf())
```

```python
mask_01 = km_0.mask_from_labels('1')
pca_01 = srtr.PCATransform(sp,mask_01,defaults.pca_trans)
pca_01.transform()
km_01 = srtr.KMeansCluster(sp,mask_01,pca_01.trnsmtrx,defaults.km_cluster)
km_01.select()
sview.plot_clusters(sp,km_01)
display(gcf())
```

```python
mask_001 = km_00.mask_from_labels('1')
md_params = {'trans_dims':1}
mdt_001 = srtr.MedTrans(sp,mask_001,md_params)
mdt_001.transform()
th_001 = srtr.ThreshSelector(sp,mask_001,mdt_001.trnsmtrx,{})
thresh = percentile(mdt_001.trnsmtrx,95)
th_001.select(thresh)
sview.plot_clusters(sp,th_001)
display(gcf())
```

```python

mask_0011 = th_001.mask_from_labels('1')
pca_0011 = srtr.PCATransform(sp,mask_0011,defaults.pca_trans)
pca_0011.transform()
km_0011 = srtr.KMeansCluster(sp,mask_0011,pca_0011.trnsmtrx,defaults.km_cluster)
km_0011.select()
sview.plot_clusters(sp,km_0011)
mask_000 = km_00.mask_from_labels('0')
display(gcf())
```
